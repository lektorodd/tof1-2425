## Pyplot

Pyplot er ein del av matplotlib-biblioteket. Me kan bruka pyplot til å laga figurar:

````python
# importere bibliotek
import matplotlib.pyplot as plt
import numpy as np

# definere variblar
x = np.linspace(-2, 2, 100)
y = x**2

# plotte
plt.plot(x, y, "r--")
plt.grid()       # rutenett på grafen
plt.axvline()    # vertikal linje (gjennom origo)
plt.axhline()    # horisontal linje (gjennom origo)
plt.show()
````

## Lagre plot

Viss du vil lagra plottet som ei fil, kan du bruka `savefig`-funksjonen. 

```python

```
Denne **må** kallast før `plt.show()`.

:::




````python
# importere bibliotek
import matplotlib.pyplot as plt
import numpy as np

# definere variblar
x = np.linspace(-2, 2, 100)
y = x**2

# plotte
plt.plot(x, y, color = "blue", linestyle = "--")

# pynte
plt.title("Plot som viser andregradsfunksjon", fontsize = 16)
plt.xlabel("x", fontsize = 12)
plt.ylabel("f(x)", fontsize = 12)
plt.grid()                                           # rutenett på grafen
plt.axvline(color = "black")                         # vertikal linje (gjennom origo)
plt.axhline(color = "black")                         # horisontal linje (gjennom origo)
#plt.savefig("plott.png")
plt.show()
````
